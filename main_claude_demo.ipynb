Using Claude's API to send a question and receive a response.

## Import Libraries

```python
import os
import anthropic
from dotenv import load_dotenv
```

## Set AI Settings and Load Enviornment Key
Configure the OpenAI image model, specify the output location, and load the secure API key from `utils/.env`. Make sure the Claude key is added there.

```python
CLAUDE_MODEL = "claude-sonnet-4-6"

load_dotenv(dotenv_path="utils/.env")
CLAUDE_API_KEY = os.getenv("CLAUDE_API_KEY")
```

## Setup Anthropic Client

Set your API key as an environment variable.

```python
client = anthropic.Anthropic(api_key=CLAUDE_API_KEY)
```

## Create a Prompt for a Response
Create a prompt to ask the Claude model. The more detailed, the better output.

### Prompt 1 (Zero Shot)

```python
PROMPT_1 = "Give me a tongue-twister"
```

### Prompt 2 (One-Shot)

```python
PROMPT_2 = """
Give me a tongue-twister that uses 'W' and ends in a question. An example is:
Peter Piper picked a peck of pickled peppers,
A peck of pickled peppers Peter Piper picked;
If Peter Piper picked a peck of pickled peppers,
Where’s the peck of pickled peppers Peter Piper picked?
"""
```

## Generate Response

This function sends a prompt to Claude then prints the response.

```python
def claude_chat(PROMPT):
    content = []
    content.append({"type": "text", "text": PROMPT})

    messages = [
        {
            "role": "user",
            "content": content
        }
    ]
    params = {
            "model": CLAUDE_MODEL,
            "max_tokens": 1024,
            "temperature": 0,
            "messages": messages,  
        }

    response = client.messages.create(**params)
    convo = response.content[0].text
    return print(convo)
```

## Run Prompt 1 (Zero Shot) Chat

Run the chat function for Zero Shot prompt.

```python
print(f"Your Prompt...\n{PROMPT_1}\n\nClaude Response...\n")
claude_chat(PROMPT_1)
```

## Run Prompt 2 (One Shot) Chat

Run the chat function for One Shot prompt.

```python
print(f"Your Prompt...\n{PROMPT_2}\n\nClaude Response...\n")
claude_chat(PROMPT_2)
```
